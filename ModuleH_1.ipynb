Section 22.1: ODE Initial Value Problem Statement

A differential equation is a function that defines the relationship between a function, its independent variable, and any number of it's derivatives.

Ordinary differential equations (abbreviated "ODE") are differential equations where the independant variable is one-dimensional while Partial differential equations are multi-dimensional differential equations. This section will focus on ODEs specifically.

In general, an ODE can be written in the form:

$F(x,f(x),\frac{df(x)}{dx},\frac{d^2f(x)}{dx^2},...,\frac{d^{n-1}f(x)}{dx^{n-1}})=\frac{d^nf(x)}{dx^n}$

A general solution is a function g(x) that satisfies the differential equation, while a particular solution is a funciton p(x) that satisfies the differential equation when provided with a set of n initial values for the function and it's derivatives.

The goal of an ODE initial value problem is to find the particular solution to an ODE when given a set of n initial values, with n representing the number of derivatives or "order" of the problem.

Section 22.2: Reduction of Order

In order to numerically solve an ODE, it is most convenient to reduce the problem to first order. This can be done by first representing the problem as a vector S(t) referred to as the state of the function. It contains all of the independant variables as a function of time. S(t) is represented as:

$S(t) = \begin{pmatrix}
f(t) \\
f^{(1)}(t) \\
f^{(2)}(t) \\
...\\
f^{n-1}(t)
\end{pmatrix}
$

The derivative of the state of a system can be expressed as the product of S(t) and an nXn matrix, which turns the nth order ODE into n 1st order ODEs that are coupled together, allowing them to be solved numerically.

Section 22.3: The Euler Method

The Explicit Euler Formula approximates the state of a function $S(t_{j+1})$ given the state $S(t_j)$ and a step size of h.

$S(j_{j+1}) = S(t_j) + h\frac{dS(t_j)}{dt}$ where $h = (t_{j+1} - t_j)$

If we are given the initial state $S(t_0)$, we can approximate the state of $S(t_f)$.

This approximation is demonstrated below using python.

```python
# The following example approximates the solution for a simple ODE and compares it to the exact solution, -e^t

import numpy as np
import matplotlib.pyplot as plt

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i]) #Application of Euler's method for each initial value in the state

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

Smaller step sizes can result in more accurate approximations. Below is an approximation using a step size of h=0.005. As you can see, this approximation is much closer to the actual solution.
